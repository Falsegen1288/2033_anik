```python
df = df.drop(['bureau_148', 'bureau_433', 'bureau_435', 'bureau_436', 'bureau_437', 'bureau_438', 'bureau_444', 'bureau_445', 'bureau_446', 'bureau_447', 'bureau_448', 'bureau_449', 'bureau_451', 'onus_attribute_43', 'onus_attribute_44', 'onus_attribute_45', 'onus_attribute_46', 'onus_attribute_47', 'onus_attribute_48'], axis = 1)
df
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Ensure reproducibility
np.random.seed(42)

# Separate columns with numerical data (autoencoders work with numerical data only)
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Scale the numerical data to [0, 1]
scaler =  MinMaxScaler()
scaled_data = scaler.fit_transform(df[numerical_columns])

# Replace NaNs with 0 temporarily for autoencoder training
# Autoencoders will learn to reconstruct the original data
nan_mask = np.isnan(scaled_data)  # Store the mask of NaN values
scaled_data[nan_mask] = 0

# Define the autoencoder
input_dim = scaled_data.shape[1]
autoencoder = Sequential([
    Dense(64, activation='relu', input_dim=input_dim),
    Dense(32, activation='relu'),
    Dense(64, activation='relu'),
    Dense(input_dim, activation='sigmoid')  # Output layer should match input dimension
])
autoencoder.compile(optimizer='adam', loss='mse')

# Train the autoencoder
autoencoder.fit(scaled_data, scaled_data, epochs=30, batch_size=32, verbose=1)

# Predict the reconstructed data
reconstructed_data = autoencoder.predict(scaled_data)

# Replace only the NaN values in the original scaled data
scaled_data[nan_mask] = reconstructed_data[nan_mask]

# Inverse transform the scaled data back to the original scale
imputed_data = scaler.inverse_transform(scaled_data)

# Update the original dataframe with imputed values
df_imputed = df.copy()
df_imputed[numerical_columns] = imputed_data

# Print a summary
print("NaN values successfully imputed using an autoencoder!")
print(df_imputed.head())
```

# Train-test Split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_imputed.drop(['account_number','bad_flag'], axis=1)
y = df_imputed['bad_flag']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the data
X_scaled = scaler.fit_transform(X)

# Convert back to a DataFrame (if X is a DataFrame)
# import pandas as pd
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)

print("Data scaled using MinMaxScaler!")
print(X_scaled)
```

```python
X_train, X_val, y_train, y_val = train_test_split(X_scaled, y, test_size = 0.2,random_state=42)
```

```python
print(X_train.shape, X_val.shape)
```

```python
X_train.describe()
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Create the LightGBM dataset
train_data = lgb.Dataset(X_train, label=y_train)
val_data = lgb.Dataset(X_val, label=y_val, reference=train_data)

# Set random hyperparameters
params = {
    'objective': 'binary',          # Binary classification
    'boosting_type': 'gbdt',        # Gradient Boosting Decision Tree
    'metric': 'binary_error',       # Metric to evaluate performance (binary classification error)
    'learning_rate': 0.01,          # Step size shrinkage
    'num_leaves': 32,               # Maximum leaves in one tree
    'max_depth': 7,                # No limit on tree depth
    'min_data_in_leaf': 20,         # Minimum number of data points in a leaf
    'feature_fraction': 0.7,        # Fraction of features used per iteration
    'bagging_fraction': 0.7,        # Subsample ratio of data
    'bagging_freq': 5,              # Bagging frequency
    'verbose': -1                   # Suppress logs
}

# Train the model
model = lgb.train(
    params,
    train_data,
    valid_sets=[train_data, val_data],  # Validation for early stopping
    num_boost_round=1000              # Maximum number of boosting rounds           # Stop if validation score doesn't improve
)

# Predict probabilities on the validation set
y_pred_prob = model.predict(X_val)

# Convert probabilities to binary (0 or 1) based on a threshold of 0.5
y_pred = (y_pred_prob >= 0.5).astype(int)

# Evaluate the model using accuracy
accuracy = accuracy_score(y_val, y_pred)
print(f"Validation Accuracy: {accuracy:.4f}")
```

```python
import lightgbm as lgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

# Create the LightGBM dataset
train_data = lgb.Dataset(X_train, label=y_train)
val_data = lgb.Dataset(X_val, label=y_val, reference=train_data)

# Set random hyperparameters
params = {
        'objective': 'binary',          # Binary classification
    'boosting_type': 'gbdt',        # Gradient Boosting Decision Tree
    'metric': 'auc',       # Metric to evaluate performance (binary classification error)
    'learning_rate': 0.05,          # Step size shrinkage
    'num_leaves': 32,               # Maximum leaves in one tree
    'max_depth': -1,                # No limit on tree depth
    'min_data_in_leaf': 20,         # Minimum number of data points in a leaf
    'feature_fraction': 0.7,        # Fraction of features used per iteration
    'bagging_fraction': 0.7,        # Subsample ratio of data
    'bagging_freq': 5,              # Bagging frequency
    'verbose': -1,                  # Suppress logs
    'random_state': 42,               # Seed for reproducibility
    'device' : 'gpu'
}



# Train the model
model = lgb.train(
    params,
    train_data,
    valid_sets=[train_data, val_data],  # Validation for early stopping
    num_boost_round=1000              # Maximum number of boosting rounds
)

# Predict probabilities on the validation set
y_pred = model.predict(X_val)

# Evaluate the model
auc_score = roc_auc_score(y_val, y_pred)
print(f"Validation AUC: {auc_score:.4f}")
```
